```python
import numpy as np
import time
```

```python
#execution performance
size=1_000_000
#time
py_list=list(range(size))
start=time.time()
sq_list=[x**2 for x in py_list]
end=time.time()
print(f"python list time {end-start} seconds")
np_arr=np.array(py_list)
start=time.time()
sq_array=np_arr**2 #vectorised operation
end=time.time()
print(f"numpy array timr {end-start} seconds")
```

```python
#memory
import sys
print(f"python list size= {sys.getsizeof(py_list)} bytes")
print(f"numpy aaray size= {np_arr.nbytes} bytes")
```
