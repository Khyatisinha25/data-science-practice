```python
import requests
from bs4 import BeautifulSoup
import os

os.makedirs("scraped_data", exist_ok=True)

# STEP 1: Download all pages
page_count = 1
while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("no valid pages anymore..")
        break

    with open(f"scraped_data/quotes{page_count}.html", "w", encoding="utf-8") as f:
        f.write(res.text)
        print(f"downloaded data from page {page_count}")

    page_count = page_count + 1

# STEP 2: Extract life quotes from all pages
life_quotes = []

for page_count in range(1, 11):
    with open(f"scraped_data/quotes{page_count}.html", "r", encoding="utf-8") as f:
        html_content = f.read()

    soup = BeautifulSoup(html_content, "lxml")
    all_quotes = soup.select("div.quote")

    for q in all_quotes:
        all_tags = []
        for tag in q.select(".tags .tag"):
            all_tags.append(tag.get_text())

        if "life" in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])

# STEP 3: Print results
print(f"\nTotal life quotes found: {len(life_quotes)}\n")
for quote in life_quotes:
    print(f"Quote  : {quote[0]}")
    print(f"Author : {quote[1]}")
    print("-" * 50)
```
