```python
#q1
import numpy as np
data=np.random.normal(loc=70,scale=10,size=200)
sns.histplot(data,kde=True)
plt.axvline(x=data.mean(),label="mean",linestyle="--",color="red")
plt.title("distribution of scores")
plt.xlabel("scores")
plt.ylabel("frequency")
plt.legend()
plt.tight_layout()
```

```python
#q2
sns.set_theme(style="darkgrid")
penguins=sns.load_dataset("penguins")
sns.scatterplot(
    data=penguins,
    x="body_mass_g",
    y="flipper_length_mm",
    hue="species"
)
plt.title("flipper length  vs body mass")
```

```python
#q3
tips=sns.load_dataset("tips")
tips.head()
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.title("total bill for each day")
```

```python
#interpretation->median highest-sat,several outliers can be seen on sat and sun
```

```python
#q4
tips=sns.load_dataset("tips")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
    
)
plt.title("total bill vs tip")
plt.xlabel("Total bill")
plt.ylabel("Tips")
```

```python
tips_pivot = tips.pivot_table(index="day", columns="time", values="tip", aggfunc="mean")
```

```python
sns.heatmap(
    tips_pivot,
    cmap="coolwarm",
    annot=True,
    
)
```
