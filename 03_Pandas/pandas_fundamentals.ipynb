```python
#melt
df=pd.DataFrame({
    "country":["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
    "tax":[20,30,40,50]
})
df
```

```python
melted_df=df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit","tax"],
    var_name="metrics",
    value_name ="value"
)
melted_df
```

```python
#PIVOT
df=pd.DataFrame({
    "country":["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
   
})
df
```

```python
melted_df=df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name ="value"
)
melted_df
```

```python
melted_df.pivot(
    index=["country","year"],
    columns="metrics",
    values="value"
)
```

```python
#merging and joining data
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})
```

```python
df_customers
```

```python
df_orders
```

```python
#inner join---> only common data
pd.merge(df_customers,df_orders ,on ="customer_id")
```

```python
#left join---> keeps df_customers values as it was left 
pd.merge(df_customers,df_orders ,on ="customer_id", how="left")
```

```python
#right join--->keeps df_orders values as it was right
pd.merge(df_customers,df_orders ,on ="customer_id", how="right")
```

```python
#outer join--->all data of both
pd.merge(df_customers,df_orders ,on ="customer_id",how="outer")
```
