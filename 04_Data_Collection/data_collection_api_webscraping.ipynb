```python
#WEB SCRAPING-->scrapethissite.com
import requests
```

```python
URL="https://www.scrapethissite.com/pages/simple/"
res=requests.get(URL)
print(res.status_code)
```

```python
if res.status_code==200:
    print(res.headers)

    
#saving html(res.text) in our scraped_data file
with open("scraped_data/data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
#BEAUTIFULSOUP
from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data1.html","r",encoding="utf-8") as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
```

```python
soup.find("h3")
```

```python
all_h3=soup.find_all("h3")
all_countries=[]
for h3 in all_h3:
    name=(h3.get_text(strip=True))
    #print(h3.find_parent("div")["class"])
    population=(h3.find_next("div").select_one("span.country-population").get_text(strip=True))
    all_countries.append([name,population])
```

```python
import pandas as pd
df=pd.DataFrame(all_countries,columns=["Name","Population"])
```

```python
df
```

```python
df.to_csv("cleaned(s)_data/data.csv",index=False)
```
