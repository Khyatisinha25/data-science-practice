# Customer Feedback Data Processing
## Python Data Cleaning & Analysis Practice

**What this covers:**
- Loading JSON data
- Cleaning messy data (text ratings, missing values, duplicates)
- Extracting insights (average rating, poor rating %)
- Building a simple recommendation system based on ratings

```python
import json
```

```python
#load the data
def load_data(filename):
    with open (filename,"r") as f:
       data=json.load(f)
    return data
```

```python
data=load_data("store.json")
print(data)
print(type(data))
```

## Step 1: Data Cleaning
Converting text ratings to numbers, handling missing ages, removing duplicate users.

```python
#clean and structure the data
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    for user in data:
        #clean ratings
        raw_rating=str(user["rating"]).strip().lower()
        if (raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
        user["rating"]=raw_rating
        #handles missing values
        raw_age=user.get("age")
        if (raw_age==None):
            user["age"]=None
        #deduplication
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])    
        cleaned_data.append(user) 
    return cleaned_data  
        
         
```

```python
data=clean_data(data)
```

## Step 2: Extracting Insights
Calculating average rating and percentage of poor ratings.

```python
#get meaningfull insights from data
def get_insights(data):
    #avg rating
    tot_rating=0
    for user in data:
        tot_rating+=float(user["rating"])
    print(f"avg rating={tot_rating/len(data)}")    

    #percentage of users with poor rating
    poor_rating=0
    for user in data:
        if (float(user["rating"])<3):
            poor_rating+=1
    print(f"% of users with poor rating={poor_rating/len(data)*100}")
    
```

```python
get_insights(data)
```

## Step 3: Recommendation Engine
Recommending phone brand based on user rating — Apple for rating ≥4, Samsung otherwise.

```python
#Build recommendation feature
def get_recommendation(data):
    recommendation=[]

    for user in data:
        curr_recomm={}
        curr_recomm["name"]=user["name"]

        if (float(user["rating"])>=4):
            curr_recomm["brand"]="Apple"
        else:
            curr_recomm["brand"]="Samsung"

        recommendation.append(curr_recomm)
        
    return recommendation

        
```

```python
get_recommendation(data)
```
